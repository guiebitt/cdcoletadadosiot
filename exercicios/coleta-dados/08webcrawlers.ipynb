# Primeiro Web Crawler

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
import datetime
import random

random.seed(datetime.datetime.now())
```

```python
def getLink(articuleUrl):
    html = urlopen('https://en.wikipedia.org{}'.format(articuleUrl)) # 'https://en.wikipedia.org/wiki/Kevin_Bacon'
    bs = BeautifulSoup(html.read(),'html.parser')
    divContent = bs.find('div', { 'id': 'bodyContent'})
    listaLinks = divContent.findAll('a', href=re.compile('^(/wiki/)((?!:).)*$')) # Tem que comeÃ§ar com /wiki/, pode ter qualquer caracter exceto :, termina com qualquer coisa
    return listaLinks
```

```python
links = getLink('/wiki/Kevin_Bacon')
while len(links) > 0:
    newArticle = links[random.randint(0, len(links) - 1)].attrs['href']
    print(newArticle)
    links = getLink(newArticle)
```

```python
for link in links:
    if 'href' in link.attrs: # and link.attrs['href'][0:1] != '#':
        print(link.attrs['href'])
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re

pages = set()
def getLinks(articleUrl):
    global pages
    html = urlopen('http://en.wikipedia.org{}'.format(articleUrl))
    bs = BeautifulSoup(html, 'html.parser')
    try:
        print(bs.h1.get_text())
        print(bs.find(id='mw-content-text').findAll('p')[0])
        print(bs.find(id='ca-edit').find('span').find('a').attrs['href'])
    except AttributeError:
        print('This page is missing something! Continuing...')
    for link in bs.findAll('a', href=re.compile('^(/wiki/)((?!:).)*$')):
        if 'href' in link.attrs:
            if link.attrs['href'] not in pages:
                #encontramos uma nova pagina
                newPage = link.attrs['href']
                print('-'*20)
                print(newPage)
                pages.add(newPage)
                getLinks(newPage)
    
getLinks('')
```
